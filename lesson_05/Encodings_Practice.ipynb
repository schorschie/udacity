To complete the second task, we need to first reshape the dataset so that all Pokémon types are recorded in a single column. This will add duplicates of Pokémon with two types, which is fine for the task to be performed.

```python
type_cols = ['type_1','type_2']
non_type_cols = pokemon.columns.difference(type_cols)
pkmn_types = pokemon.melt(id_vars = non_type_cols, value_vars = type_cols, 
                          var_name = 'type_level', value_name = 'type').dropna()
pkmn_types.head()
```

**Task 2**: How do weights and heights compare between Fairy type Pokémon and Dragon type Pokémon? You may want to subset your dataframe before proceeding with the plotting code. **Hint**: If you remember from the univariate plots lesson, one of your axis variables may need to be transformed. If you plan on using FacetGrid, its `.set()` method will be vital for adjusting the axis scaling and tick marks.
```python
# YOUR CODE HERE
plt.scatter(x='weight', y='height', 
            data=pkmn_types.query("type == 'fairy'"),
            color='red', label='Fairy');
plt.scatter(x='weight', y='height', 
            data=pkmn_types.query("type == 'dragon'"),
            color='blue', label='Dragon');
plt.legend();
plt.xscale('log')
```
